\begin{equation*}
 \begin{cases}
   \dot{x}_{i} = y_{i} - ax^{3}_{i} + bx^{2}_{i} - z_{i} + I - k_{i}(x_{i} - v_{s})\Gamma(x_{j});\\
   \dot{y}_{i} = c - dx^{2}_{i} - y_{i};\\
   \dot{z}_{i} = r(s(x_{i}-x_{r}) - z_{i}).\\
 \end{cases}
\end{equation*}

$$\Gamma(x) = \frac{1}{1 + \exp^{-\lambda(x - \Theta)}}$$

```python
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

 $Параметры:\\
 a = 1,\ b = 3,\ c = 1,\ d = 5,\ x_{r} = -1.6,\ r = 0.01,\ s = 5,\ I = 4,\ v_{s} = 2,\ \lambda = 10,\ \Theta = -0.25$

```python
a = 1
b = 3
c = 1
d = 5
x_r = -1.6
r = 0.01
s = 5
I = 4
v_s = 2
lambd = 10
Theta = -0.25
```

```python
Gamma = lambda x_i: 1 / (1 + np.exp(-lambd * (x_i - Theta)))
```

```python
def two_HRs(t, X, k1, k2):
    x1, y1, z1, x2, y2, z2  = X
    dx1 = y1 - a*x1**3 + b*x1**2 -z1 + I - k1*(x1 - v_s) * Gamma(x2)
    dy1 = c - d*x1**2 - y1
    dz1 = r*(s*(x1 - x_r) - z1)
    
    dx2 = y2 - a*x2**3 + b*x2**2 -z2 + I - k2*(x2 - v_s) * Gamma(x1)
    dy2 = c - d*x2**2 - y2
    dz2 = r*(s*(x2 - x_r) - z2)
    return [dx1, dy1, dz1, dx2, dy2, dz2]
```

```python
#a)
# k1 = 0.04
# k2 = 0.04

initials = [-1.5, 0, 0, -1.5, 0, 0]

sol = solve_ivp(two_HRs, [0, 5000], initials, rtol = 1e-11, atol = 1e-11, dense_output=True, args = (0.04, 0.04))
ts = sol.t
x1s, y1s, z1s, x2s, y2s, z2s = sol.y
```
