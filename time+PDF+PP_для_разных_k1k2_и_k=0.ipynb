```python
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.signal import argrelextrema
```

```python
#задаём параметры, функцию гамма и саму систему из двух связных нейронов
a = 1
b = 3
c = 1
d = 5
x_r = -1.6
r = 0.01
s = 5
I = 4
v_s = 2
lambd = 10
Theta = -0.25

Gamma = lambda x_i: 1 / (1 + np.exp(-lambd * (x_i - Theta)))

def two_HRs(t, X, k1, k2, k):
    x1, y1, z1, x2, y2, z2  = X
    dx1 = y1 - a*x1**3 + b*x1**2 -z1 + I - k1*(x1 - v_s) * Gamma(x2) + k * (x2 - x1)
    dy1 = c - d*x1**2 - y1
    dz1 = r*(s*(x1 - x_r) - z1)
    
    dx2 = y2 - a*x2**3 + b*x2**2 -z2 + I - k2*(x2 - v_s) * Gamma(x1) + k * (x1 - x2)
    dy2 = c - d*x2**2 - y2
    dz2 = r*(s*(x2 - x_r) - z2)
    return [dx1, dy1, dz1, dx2, dy2, dz2]
```

```python
###1
k1 = 0.04
k2 = 0.04
k = 0.0

initials = [-1.5, 0, 0, -1.5, 0, 0]

sol1 = solve_ivp(two_HRs, [0, 20000], initials, rtol = 1e-11, atol = 1e-11, dense_output=True, args = (k1, k2, k))
ts1 = sol1.t
x1s1, y1s1, z1s1, x2s1, y2s1, z2s1 = sol1.y

##2

k1 = 0.0532
k2 = 0.0532
k = 0.0

initials = [x1s1[-1], y1s1[-1], z1s1[-1], x2s1[-1], y2s1[-1], z2s1[-1]]

sol2 = solve_ivp(two_HRs, [0, 50000], initials, rtol = 1e-11, atol = 1e-11, dense_output=True, args = (k1, k2, k))
ts2 = sol2.t
x1s2, y1s2, z1s2, x2s2, y2s2, z2s2 = sol2.y

###3

k1 = 0.07
k2 = 0.07
k = 0.0

initials = [x1s2[-1], y1s2[-1], z1s2[-1], x2s2[-1], y2s2[-1], z2s2[-1]]

sol3 = solve_ivp(two_HRs, [0, 20000], initials, rtol = 1e-11, atol = 1e-11, dense_output=True, args = (k1, k2, k))
ts3 = sol3.t
x1s3, y1s3, z1s3, x2s3, y2s3, z2s3 = sol3.y

###4

k1 = -0.04
k2 = -0.04
k = 0.0

initials = [-1.5, 0, 0, -1.5, 0, 0]

sol4 = solve_ivp(two_HRs, [0, 15000], initials, rtol = 1e-11, atol = 1e-11, dense_output=True, args = (k1, k2, k))
ts4 = sol4.t
x1s4, y1s4, z1s4, x2s4, y2s4, z2s4 = sol4.y

###5

k1 = -0.06
k2 = -0.06
k = 0.0

initials = [x1s4[-1], y1s4[-1], z1s4[-1], x2s4[-1], y2s4[-1], z2s4[-1]]

sol5 = solve_ivp(two_HRs, [0, 20000], initials, rtol = 1e-11, atol = 1e-11, dense_output=True, args = (k1, k2, k))
ts5 = sol5.t
x1s5, y1s5, z1s5, x2s5, y2s5, z2s5 = sol5.y

###6

k1 = -0.07
k2 = -0.07
k = 0.0

initials = [x1s5[-1], y1s5[-1], z1s5[-1], x2s5[-1], y2s5[-1], z2s5[-1]]

sol6 = solve_ivp(two_HRs, [0, 20000], initials, rtol = 1e-11, atol = 1e-11, dense_output=True, args = (k1, k2, k))
ts6 = sol6.t
x1s6, y1s6, z1s6, x2s6, y2s6, z2s6 = sol6.y
```

```python
spikes_ind1 = argrelextrema(xsum1, np.greater) 
spikes1 = xsum1[spikes_ind1]

spikes_ind2 = argrelextrema(xsum2, np.greater) 
spikes2 = xsum2[spikes_ind2]

spikes_ind3 = argrelextrema(xsum3, np.greater) 
spikes3 = xsum3[spikes_ind3]

spikes_ind4 = argrelextrema(xsum4, np.greater) 
spikes4 = xsum4[spikes_ind4]

spikes_ind5 = argrelextrema(xsum5, np.greater) 
spikes5 = xsum5[spikes_ind5]

spikes_ind6 = argrelextrema(xsum6, np.greater) 
spikes6 = xsum6[spikes_ind6]

mu1 = np.mean(spikes1)
mu2 = np.mean(spikes2)
mu3 = np.mean(spikes3)
mu4 = np.mean(spikes4)
mu5 = np.mean(spikes5)
mu6 = np.mean(spikes6)

sigma1 = np.std(spikes1)
sigma2 = np.std(spikes2)
sigma3 = np.std(spikes3)
sigma4 = np.std(spikes4)
sigma5 = np.std(spikes5)
sigma6 = np.std(spikes6)

Hs1 = mu1 + 6*sigma1
Hs2 = mu2 + 6*sigma2
Hs3 = mu3 + 6*sigma3
Hs4 = mu4 + 6*sigma4
Hs5 = mu5 + 6*sigma5
Hs6 = mu6 + 6*sigma6

print("Hs1  = ", Hs1)
print("Hs2  = ", Hs2)
print("Hs3  = ", Hs3)
print("Hs4  = ", Hs4)
print("Hs5  = ", Hs5)
print("Hs6 = ", Hs6)
```
