\begin{equation*}
 \begin{cases}
   \dot{x} = y - ax^{3} + bx^{2} + I - z;\\
   \dot{y} = c - dx^{2} - y;\\
   \dot{z} = r(s(x-x_{r}) - z).\\
 \end{cases}
\end{equation*}


```python
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import sympy as sp
sp.init_printing()
```

 $Константы:\\
 a = 1,\\
 c = 1\\
 d = 5\\
 x_{rest} = -1.6\ (устанавливает\ потенциал\ покоя\ системы)$

```python
a = 1
c = 1
d = 5
x_rest = -1.6
```

$Бифуркационные\ параметры:\\
b - позволяет\ переключаться\ между\ спайковыми\ и\ бёрстовыми\ режимами\ и\ контролировать\ частоту\ спайков,\\ 
r - контролирует\ скорость\ изменения\ медленной\ переменной\ z,\\ 
s - управляет\ адаптацией,\\
I - внешний\ ток$

$1.\ Нейрон\ в\ состоянии\ покоя\ (только\ одно\ состояние\ равновесия)\\
b = 0.5\\ 
r = 0.01\\ 
s = 4\\
I = 1.5$

```python
b = 0.5
r = 0.01
s = 4
I = 1.5
```

```python
def HR3(t, X):
    x, y, z = X
    return [y - a*x**3 + b*x**2 + I - z, c - d*x**2 - y, r*(s*(x - x_rest) - z)]
```

```python
sol = solve_ivp(HR3, [0, 1000], [1, 1, 1], rtol = 1e-11, atol = 1e-11, dense_output=True)
```

```python
ts = sol.t
xs, ys, zs = sol.y
```

```python
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(projection='3d')
ax.plot(xs, ys, zs)
ax.scatter(xs[0], ys[0], zs[0], c='g', label='Начало')
ax.scatter(xs[-1], ys[-1], zs[-1], c='r', label='Конец')
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")

fig.savefig('ppb05i15')
```

```python
fig = plt.figure()

plt.plot(ts, xs)
plt.xlabel(r'$t$')
plt.ylabel(r'$x$')

fig.savefig('tdb05i15')
```
